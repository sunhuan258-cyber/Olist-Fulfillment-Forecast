```python
import pandas as pd
df_orders = pd.read_csv(
    "C:\\Users\\Administrator\\Desktop\\OLIST!\\olist_orders_dataset.csv",
    parse_dates=[
        'order_purchase_timestamp',
        'order_approved_at',
        'order_delivered_carrier_date',
        'order_delivered_customer_date',
        'order_estimated_delivery_date'])
df_payments = pd.read_csv(
    "C:\\Users\\Administrator\\Desktop\\OLIST!\\olist_order_payments_dataset.csv",
    dtype={'payment_sequential': 'object', 'payment_installments': 'object'})
df_order_payments = pd.merge(
    left=df_orders,
    right=df_payments,
    on='order_id',
    how='left')
print("--- “订单-支付”连接后，宏观鸟瞰报告 ---")
print(f"连接前的订单总表行数: {len(df_orders)}")
print(f"连接后的宽表行数: {len(df_order_payments)}")
print("\n连接后，数据预览:")
print(df_order_payments.head())
print("\n连接后，数据信息:")
df_order_payments.info()
```

```python
import pandas as pd
df_items = pd.read_csv(
    "C:\\Users\\Administrator\\Desktop\\OLIST!\\olist_order_items_dataset.csv",
    dtype={'order_item_id': 'object'})
df_order_items_payments = pd.merge(
    left=df_order_payments,
    right=df_items,
    on='order_id',
    how='left')
print("--- “订单-支付-商品”连接后，宏观鸟瞰报告 ---")
print(f"连接前的宽表行数: {len(df_order_payments)}")
print(f"连接后的超级宽表行数: {len(df_order_items_payments)}")
print("\n连接后，数据预览:")
print(df_order_items_payments.head())
print("\n连接后，数据信息:")
df_order_items_payments.info()
    
    
```

```python
import pandas as pd
df_customers = pd.read_csv(
    "C:\\Users\\Administrator\\Desktop\\OLIST!\\olist_customers_dataset.csv",
    dtype={'customer_zip_code_prefix': 'object'})
df_final_temp = pd.merge(
    left=df_order_items_payments,
    right=df_customers,
    on='customer_id',
    how='left')
print("--- “订单-商品-支付-客户”连接后，宏观鸟瞰报告 ---")
print(f"连接前的超级宽表行数: {len(df_order_items_payments)}")
print(f"连接后的究极宽表行数: {len(df_final_temp)}")
print("\n连接后，数据预览:")
print(df_final_temp.head())
print("\n连接后，数据信息:")
df_final_temp.info()
```

```python
import pandas as pd
df_products = pd.read_csv(
    "C:\\Users\\Administrator\\Desktop\\OLIST!\\olist_products_dataset.csv")
df_sellers = pd.read_csv(
    "C:\\Users\\Administrator\\Desktop\\OLIST!\\olist_sellers_dataset.csv",
    dtype={'seller_zip_code_prefix': 'object'})
df_with_products = pd.merge(
    left=df_final_temp,
    right=df_products,
    on='product_id',
    how='left')
df_with_sellers = pd.merge(
    left=df_with_products,
    right=df_sellers,
    on='seller_id',
    how='left')
print("--- “订单-商品-支付-客户-产品-卖家”连接后，宏观鸟瞰报告 ---")
print(f"连接前的究极宽表行数: {len(df_final_temp)}")
print(f"连接后的最终宽表行数: {len(df_with_sellers)}")
print("\n连接后，数据预览:")
print(f"最终宽表维度: {df_with_sellers.shape}")
print(df_with_sellers.head())
print("\n连接后，数据信息:")
df_with_sellers.info()
```

```python
import pandas as pd
df_translation = pd.read_csv(
    "C:\\Users\\Administrator\\Desktop\\OLIST!\\product_category_name_translation.csv")
df_geolocation_raw = pd.read_csv(
    "C:\\Users\\Administrator\\Desktop\\OLIST!\\olist_geolocation_dataset.csv",
    dtype={'geolocation_zip_code_prefix': 'object'})
df_geolocation = df_geolocation_raw.groupby('geolocation_zip_code_prefix').agg({
    'geolocation_lat': 'mean',
    'geolocation_lng': 'mean'}).reset_index()
df_with_translation = pd.merge(
    left=df_with_sellers,
    right=df_translation,
    on='product_category_name',
    how='left')
df_with_customer_geo = pd.merge(
    left=df_with_translation,
    right=df_geolocation,
    left_on='customer_zip_code_prefix',
    right_on='geolocation_zip_code_prefix',
    how='left')
df_final = pd.merge(
    left=df_with_customer_geo,
    right=df_geolocation,
    left_on='seller_zip_code_prefix',
    right_on='geolocation_zip_code_prefix',
    how='left',
    suffixes=('_customer', '_seller'))
print("--- “最终分析宽表”构建完毕！宏观鸟瞰报告 ---")
print(f"最终宽表的维度: {df_final.shape}")
print("\n最终宽表，数据信息:")
print(df_final[['product_category_name_english', 'geolocation_lat_customer',
                'geolocation_lng_customer', 'geolocation_lat_seller', 'geolocation_lng_seller']].info())
                
```

```python
final_output_path = "C:\\Users\\Administrator\\Desktop\\OLIST!\\olist_final_analysis_dataset.csv"
df_final.to_csv(final_output_path, index=False)
print(f"--- “数据泰坦”已成功实体化！---")
print(f"我们最宝贵的核心资产，已安全地，存储于：")
print(final_output_path)
```
